```python
driver.get('https://stock-pictures.netlify.app/')
```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
soup
```

```python
img = []
for i in soup.find_all('img',class_ = 'source-img'):
    image = i.get('src')
    img.append(image)
```

```python
t= []
for i in soup.find_all('span',class_ = 'tag-color'):
    tags = i.text.split(' - ')[1]
    t.append(tags)
```

```python
l = []
for i in soup.find_all('div',class_ = 'likes-comments'):
    likes_comments = i.text
    l.append(likes_comments)
```

```python
data = []
for i in range(len(img)):
    data.append([img[i],t[i],l[i]])
print(len(data))
print(data)
```

```python
df = pd.DataFrame(data,columns = ['Image_link','Image_tags','no_of_likes_comments'])
df
```

```python
df.to_csv('images.csv', index = False)
```
